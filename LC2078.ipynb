```python
def maxDistance( colors):
        """
        :type colors: List[int]
        :rtype: int
        """
        fst={}
        lst={}
        for index,color in enumerate(colors):
            # print("C",color,"i",index)
            if color not in fst:
                fst[color]=index

            lst[color]=index
        print("fst",fst,"lst",lst)

        dist=[]
        for f_index,f_color in enumerate(fst):
          for l_indxe,l_color in enumerate(lst):
            if f_color!=l_color:
              dist.append(abs(fst[f_color]-lst[l_color]))
        print("dist",dist)
        return max(dist) if dist else -1
```
