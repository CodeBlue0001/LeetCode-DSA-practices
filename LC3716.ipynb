```python
def minMirrorPairDistence(nums):
    # check if the palindrom of the number is present or not
    result=99999
    for i in range(len(nums)):
        item=nums[i]
        rev=int(str(item)[::-1])

        if rev in nums[i+1:]:
            print("item",item,"rev",rev,"rest",nums[i+1:])
            ri=nums[i+1:].index(rev)+i+1
            ii=nums.index(item)
            print("revIdx",ri,"itemIdx",ii)
            if abs(ri-ii)<result:
                result=abs(ri-ii)
    if result==99999:
        return -1
    return result
r=minMirrorPairDistence([7,21,39,37,47,72,12,74])
print(r)
```
